```
#To replicate the working of the Numpy Linspace
import math
#Taking Higher, Lower and Number of values to be generated from the user.
# hi=input('Enter the Higher value: ')
# print(hi)
# lo=input('Enter the Lower value: ')
# print(lo)
# no=input('Enter the number of values to generated: ')
# print(no)
def Linspace(hi,lo,no):
    di=(hi-lo)/(no-1)
    an=[]
    for i in range(no):
        an.append(lo)
        lo=lo+di
    return an
Linspace(10,30,15)
```
